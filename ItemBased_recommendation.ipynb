```
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```
ml_data = pd.read_csv("u.data", sep= '\t', names=['user_id','item_id','rating','timestamp'])
```

```
ml_data.head()
```

```
ml_data.describe()
```

```
ml_rating = pd.DataFrame(ml_data.groupby('item_id')['rating'].mean())
```

```
ml_rating.head()
```

```
ml_rating['no of rating'] = ml_data.groupby('item_id')['rating'].count()
ml_rating.head()
```

```
ml_matrix = ml_data.pivot_table(index='user_id', columns='item_id', values='rating')
ml_matrix.head()
```

```
ml_rating.sort_values('no of rating', ascending=False).head(10)
```

```
ml_100 = ml_matrix[100]
```

```
ml_100
```

```
similar_to_ml_100 = ml_matrix.corrwith(ml_100)
```

```
similar_to_ml_100.head()
```

```
corr_100 = pd.DataFrame(similar_to_ml_100,columns=['Correlation'])
corr_100.dropna(inplace=True)
corr_100.head()
```

```
corr_100 = corr_100.join(ml_rating['no of rating'])
corr_100.head()
```

```
corr_100[corr_100['no of rating']>50].sort_values(by='Correlation', ascending=False).head()
```
